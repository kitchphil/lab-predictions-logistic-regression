```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
import getpass  # To get the password without showing the input
import re
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt
```

```python
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
```

```python
#SQL Query to Extract relevent columns

film =  '''
SELECT 
    t1.film_id 
    , COUNT(t3.rental_id) rented_times
    , t1.rental_duration
    , t5.name as category
    , t1.length
    , t1.rating
    , t1.special_features 
    , t1.rental_rate
    , t6.name AS language
    , CASE
    WHEN t3.rental_date BETWEEN '2005-05-01' AND '2005-05-31' THEN True
    ELSE False 
    END AS test
FROM 
    sakila.film AS t1
LEFT JOIN inventory AS t2 ON t1.film_id = t2.film_id
INNER JOIN sakila.rental AS t3 ON t2.inventory_id = t3.inventory_id
INNER JOIN sakila.film_category AS t4 ON t1.film_id = t4.film_id
INNER JOIN sakila.category AS t5 ON t4.category_id = t5.category_id
INNER JOIN sakila.language AS t6 ON t1.language_id = t6.language_id
#Where 
#    t3.rental_date Between '2005-01-01' AND '2005-12-31'
GROUP BY 
    film_id
    , t1.rental_duration
    , category
    , t1.length
    , t1.rating
    , test
    , t1.rental_rate
    , t1.special_features
    , language
; 

'''


dfr = pd.read_sql_query(film, engine)
```

```python
#loop using Regex to extract values and insert them into a list, finally putting them into their own columns


la = []
lb = []
lc = []
ld = []

for i in dfr['special_features']:
        if re.findall('Trailers', i):
            la.append(1)
        else:
            la.append(0)
dfr['trailers'] = la

for i in dfr['special_features']:
        if re.findall('Behind the Scenes', i):
            lb.append(1)
        else:
            lb.append(0)
dfr['behind the scenes'] = lb            

            
for i in dfr['special_features']:
        if re.findall('Commentaries', i):
            lc.append(1)
        else:
            lc.append(0)
dfr['commentaries'] = lc
            
            
for i in dfr['special_features']:
        if re.findall('Deleted Scenes', i):
            ld.append(1)
        else:
            ld.append(0)
dfr['deleted scenes'] = lc

dfr = dfr.drop(['special_features'], axis=1)
```

```python
dfr
```

```python
#Conversion of the different datatypes

dfr['film_id'] = dfr['film_id'].astype(object)
dfr['rental_duration'] = dfr['rental_duration'].astype(object)
dfr['test'] = dfr['test'].astype(bool)
dfr['rental_rate'] = dfr['rental_rate'].astype(object)
dfr['trailers'] = dfr['trailers'].astype(bool)
dfr['behind the scenes'] = dfr['behind the scenes'].astype(bool)
dfr['commentaries'] = dfr['commentaries'].astype(bool)
dfr['deleted scenes'] = dfr['deleted scenes'].astype(bool)
```

```python
#Now it's time to split the x and y to train and test.

y = dfr['test']
X = dfr.drop('test', axis=1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1337)
```

```python
#Extract numerical columns
X_train_num = X_train.select_dtypes(include = np.number)

#Normalise the columns to be between 0 and 1
transformer = MinMaxScaler().fit(X_train_num) # need to keep transformer
X_train_normalized = transformer.transform(X_train_num)
X_train_norm = pd.DataFrame(X_train_normalized)
```

```python
X_train_norm.columns = X_train_num.columns
X_train_norm.head()
```

```python
#Extract categorical columns
X_train_cat = X_train.select_dtypes(include = np.object)
X_train_cat
```

```python
#Get dummies of categorical columns
X_train_categorical = X_train.select_dtypes(include = np.object)
X_train_cat = pd.get_dummies(X_train_categorical, 
                             columns=['category', 'rating','rental_duration','rental_rate','language'],
                             drop_first=True)
X_train_cat.head()
```

```python
#Concatenate the numerical and non-numerical
X_train_transformed = np.concatenate([X_train_norm, X_train_cat], axis=1)
```

```python
#Run the Train Model
from sklearn.linear_model import LogisticRegression
classification = LogisticRegression(random_state=0, solver='lbfgs',
                  multi_class='multinomial').fit(X_train_transformed, y_train)
```

```python
#Repeat the same process for Test

#Extract numerical columns
X_test_num = X_test.select_dtypes(include = np.number)

#Normalise the columns to be between 0 and 1
X_test_normalized = transformer.transform(X_test_num)
X_test_norm = pd.DataFrame(X_test_normalized)
```

```python
#Extract categorical columns
X_test_categorical = X_test.select_dtypes(include = np.object)

#Get dummies of categorical columns
X_test_cat = pd.get_dummies(X_test_categorical, 
                            columns=['category', 'rating','rental_duration','rental_rate','language'],
                            drop_first=True)

#  Make sure the columns are in the same order as Train
display(list(zip(list(X_train_cat.columns),list(X_test_cat.columns))))
X_test_cat.head()
```

```python
#Concatenate the numerical and non-numerical
X_test_transformed = np.concatenate([X_test_norm, X_test_cat], axis=1)
```

```python
# The prediction:

predictions = classification.predict(X_test_transformed)
classification.score(X_test_transformed, y_test)
```
